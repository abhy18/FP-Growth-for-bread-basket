```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/bread basket.csv')
df
```

```python
grp = df.groupby('Transaction')['Item'].apply(list).reset_index()
expand = grp['Item']
expand
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
tee = te.fit_transform(expand)
df = pd.DataFrame(tee,columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import fpgrowth
fpgrowth(df,min_support = 0.0003)
```
